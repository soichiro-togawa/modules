```
%%time
#パイソン
def fib_python(n):
    a, b = 0, 1
    for i in range(n):
        a, b = a + b, a
    return a
```

```
%%cython
def fib_cython(n):
    a, b = 0, 1
    for i in range(n):
        a, b = a + b, a
    return a
```

```
%%cython
def fib_all_typed_cython(int n):
    cdef int a,b,i
    
    a, b = 0, 1
    for i in range(n):
        a, b = a + b, a
    return a
```
